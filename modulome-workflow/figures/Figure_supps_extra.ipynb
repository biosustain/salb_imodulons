## Setup

```python
from os import path
import shutil
import seaborn as sns
import matplotlib.pyplot as plt

from pymodulon.io import load_json_model
from pymodulon.plotting import *
```

### Set plotting style

```python
sns.set_style('ticks')
plt.style.use('custom.mplstyle')
```

### Load data

```python
figure_dir = '../../reports/figures/modulome/supplementary'
data_dir = path.join('../..','data','processed', 'modulome')
```

```python
data_file = path.join(data_dir,'salb.json.gz')
ica_data = load_json_model(data_file)
```

```python
plot_explained_variance(ica_data)
plt.savefig(path.join(figure_dir,'FigS2_explained_variance.pdf'))
```

# Numbers for manuscript

```python
print('Number of iModulons:',len(ica_data.imodulon_names))
```

```python
# Calculate the sum of the explained_variance column in ica_data.imodulon_table
total_variance = ica_data.imodulon_table['explained_variance'].sum()
print('Total variance explained:',total_variance)
```

```python
#Load the E. coli precise1k ica object
precise1k = load_json_model('../../data/external/modulome/ica_objects/precise1k.json')

# Load the bbh dataframe
bbh = pd.read_csv('../../data/external/modulome/compare_ica_multi/bbh/dataframes/salb_prot_vs_ecol_prot_parsed.csv',index_col=0)
```

```python
# Convert the page size from mm to inches
width = 85 / 25.4  # ~6.69 inches
height = 80 / 25.4  # ~3.94 inches

fig, ax = plt.subplots(figsize=(width, height))

# Create a dictionary where the keys are the sample IDs and the values are the project names
groups = ica_data.sample_table['project'].to_dict()

# Generate 16 colors from the 'tab20' colormap
colors = plt.cm.tab20(range(16))

# Call the compare_gene_weights function with the specified parameters
compare_gene_weights(ica_data, 'Translation',
                     ica_data2=precise1k,
                     imodulon2='Translation',
                     ortho_file=bbh,
                     use_org1_names=False,
                     scatter_kwargs={'s': 20},
                     show_labels=True,
                     label_font_kwargs={'fontsize': 6},
                     ax_font_kwargs={'fontsize': 6},
                     ax=ax)

# Increase x and y axis text size
ax.tick_params(axis='both', which='major', labelsize=6)

# Increase the x and y-axis label size
ax.set_xlabel('Translation (S. albidoflavus)', fontsize=6)
ax.set_ylabel('Translation (E. coli)', fontsize=6)

# Move the legend to the top right outside the plot
plt.legend(loc='upper right', bbox_to_anchor=(2, 1))

# Save the figure
plt.savefig(path.join(figure_dir, 'translation_compare_weights.pdf'), dpi=300)

# Show the plot
plt.show()
```
